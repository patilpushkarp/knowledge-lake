# India vs China: Diverging Stock Market Stories

```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

# Define tickers (Yahoo Finance symbols)
india_ticker = '^NSEI'       # Nifty 50
china_ticker = '000001.SS'   # Shanghai Composite

# Download daily close data
start_date = '2024-01-01'
end_date   = '2025-11-01'

df_india = yf.download(india_ticker, start=start_date, end=end_date, progress=False)[['Close']]
df_china = yf.download(china_ticker, start=start_date, end=end_date, progress=False)[['Close']]

# Rename columns for clarity
df_india.rename(columns={'Close': 'India (Nifty 50)'}, inplace=True)
df_china.rename(columns={'Close': 'China (Shanghai Composite)'}, inplace=True)

# Merge on date
df = pd.merge(df_india, df_china, left_index=True, right_index=True, how='inner')

# Normalize both series to 100 at the start
df = df / df.iloc[0] * 100

# Plot
plt.figure(figsize=(10,5))
plt.plot(df.index, df['India (Nifty 50)'], label='India  Nifty 50', linewidth=2)
plt.plot(df.index, df['China (Shanghai Composite)'], label='China  Shanghai Composite', linewidth=2, linestyle='--')
plt.title('India vs China: Diverging Stock Market Performance (20242025)')
plt.xlabel('Date')
plt.ylabel('Index (Normalized to 100)')
plt.legend()
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```

The two Asian giants have switched roles in 2025s market narrative. From early 2024 to late 2025, Indias Nifty 50 advanced about 20%, powered by strong GDP growth, infrastructure spending, and steady foreign inflows.
Chinas Shanghai Composite, however, lagged through most of 2024 amid property market weakness and muted consumer demand and then staged a surprising recovery in 2025, climbing over 30% as stimulus measures and tech-sector easing restored confidence.
The chart captures this evolving tug-of-war: Indias consistent climb versus Chinas late but sharp rebound.

**Takeaway**: What began as a one-sided story of Indias resilience now looks like a regional rebalancing. The next phase will hinge on whether Chinas policy momentum can outlast Indias structural growth engine.
